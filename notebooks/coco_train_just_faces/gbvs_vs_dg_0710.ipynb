### Get Respective Arrays

```python
def split_arrays_for_plot(df):
    df = df.transpose()
    gbvs = df[0]
    gbvs_f03 = df[0.3]
    gbvs_f05 = df[0.5]
    gbvs_f07 = df[0.7]
    
    return gbvs, gbvs_f03, gbvs_f05, gbvs_f07
```

```python
auc_gbvs_m, auc_gbvs_f03_m, auc_gbvs_f05_m, auc_gbvs_f07_m = split_arrays_for_plot(gbvs_auc_mean)
auc_gbvs_s, auc_gbvs_f03_s, auc_gbvs_f05_s, auc_gbvs_f07_s = split_arrays_for_plot(gbvs_auc_std)

nss_gbvs_m, nss_gbvs_f03_m, nss_gbvs_f05_m, nss_gbvs_f07_m = split_arrays_for_plot(gbvs_nss_mean)
nss_gbvs_s, nss_gbvs_f03_s, nss_gbvs_f05_s, nss_gbvs_f07_s = split_arrays_for_plot(gbvs_nss_std)

sim_gbvs_m, sim_gbvs_f03_m, sim_gbvs_f05_m, sim_gbvs_f07_m = split_arrays_for_plot(gbvs_sim_mean)
sim_gbvs_s, sim_gbvs_f03_s, sim_gbvs_f05_s, sim_gbvs_f07_s = split_arrays_for_plot(gbvs_sim_std)

ig_gbvs_m, ig_gbvs_f03_m, ig_gbvs_f05_m, ig_gbvs_f07_m = split_arrays_for_plot(gbvs_ig_mean)
ig_gbvs_s, ig_gbvs_f03_s, ig_gbvs_f05_s, ig_gbvs_f07_s = split_arrays_for_plot(gbvs_ig_std)
```

### Make Plots

```python
t = np.arange(0,6)
d = 1.1
m = 8
```

```python
plt.subplots(figsize=(12,8))

plt.plot(auc_gbvs_m, '-o', label='gbvs', linewidth=d, markersize=m, c='b');
plt.errorbar(t, auc_gbvs_m, auc_gbvs_s, linestyle='None', c='b', label=None, alpha=0.45)
plt.fill_between(t, auc_gbvs_m - auc_gbvs_s, auc_gbvs_m + auc_gbvs_s, color='b', alpha=0.09)

plt.plot(auc_gbvs_f07_m, '-o', label='gbvs_f07', linewidth=d, markersize=m, c='g');
plt.errorbar(t, auc_gbvs_f07_m, auc_gbvs_f07_s, linestyle='None', c='g', label=None, alpha=0.45)
plt.fill_between(t, auc_gbvs_f07_m - auc_gbvs_f07_s, auc_gbvs_f07_m + auc_gbvs_f07_s, color='g', alpha=0.09)

plt.plot(auc_dg_m, '-o', label='dg', linewidth=d, markersize=m, c='r');
plt.errorbar(t, auc_dg_m, auc_dg_s, linestyle='None', c='r', label=None, alpha=0.45)
plt.fill_between(t, auc_dg_m - auc_dg_s, auc_dg_m + auc_dg_s, color='b', alpha=0.09)

plt.plot(auc_covsal, '-o', label='covsal', alpha=0.5)
plt.plot(auc_icf, '-o', label='icf', alpha=0.5)

plt.grid()
plt.yticks(np.arange(0.6, 0.95, step=0.05))
plt.ylim(0.61, 0.96)
plt.ylabel('score')
plt.title('AUC')
plt.xlabel('context')
plt.xticks(rotation=50)

leg = plt.legend(loc=0, ncol=3, prop=font_prop)
leg.get_frame().set_linewidth(0.1)

plt.savefig('../plots/AUC_gbvs_performance.jpg', bbox_inches='tight', pad_inches=0)
```
