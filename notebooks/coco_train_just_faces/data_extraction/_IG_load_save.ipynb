## Calculate means

```python
A = IG_mean_noface.values
A = A[:-1,:]
B = IG_mean_face_03.values
C = IG_mean_face_05.values
D = IG_mean_face_07.values
```

```python
X = np.dstack([A, B, C, D])
```

### Mean over contexts

```python
mean_models = (X.mean(axis=1))

IG_mean_models = pd.DataFrame(data=mean_models, columns=[0, 0.3, 0.5, 0.7], index=labels)
IG_mean_models
```

### Mean over models

```python
mean_cxt = (X.mean(axis=0))

IG_mean_cxt = pd.DataFrame(data=mean_cxt, columns=[0, 0.3, 0.5, 0.7], index=cxt)
IG_mean_cxt
```
