# Visualize and Analyze the data

```python
import os
os.chdir('../saliency_model/')

import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
(summary, context, model) = pickle.load(open( "19-01-31_results.p", "rb"))
```

```python
context = np.array(context)
model = np.array(model)
```

```python
nss_table = summary[:,:,0]
print(np.shape(nss_table))
sim_table = summary[:,:,1]
ig_table = summary[:,:,2]
auc_table = summary[:,:,3]
```

```python
plt.plot(nss_table)
plt.xlim(0,4)
plt.xticks([0,1,2,3,4], context)
plt.title("NSS for all models")
plt.legend(model, bbox_to_anchor = (1.1,1.05));
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5), sharey='row')

ax[0].plot(nss_table[:, [0,2,4,7,9,11,13,6]])
ax[0].set_xlim([0,6])
ax[0].set_xticks([0,1,2,3,4])
ax[0].set_xticklabels(context)
ax[0].set_title("NSS for all models without face feature")
ax[0].legend([model[i] for i in [0,2,4,7,9,11,13,6]], loc='best');

ax[1].plot(nss_table[:, [1,3,5,8,10,12,14]])
ax[1].set_xlim([0,6])
ax[1].set_xticks([0,1,2,3,4])
ax[1].set_xticklabels(context)
ax[1].set_title("NSS for all models with face feature")
ax[1].legend([model[i] for i in [1,3,5,8,10,12,14]], loc='best');
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5), sharey='row')

ax[0].plot(sim_table[:, [0,2,4,7,9,11,13,6]])
ax[0].set_xlim([0,6])
ax[0].set_xticks([0,1,2,3,4])
ax[0].set_xticklabels(context)
ax[0].set_title("SIM for all models without face feature")
ax[0].legend([model[i] for i in [0,2,4,7,9,11,13,6]], loc='best');

ax[1].plot(sim_table[:, [1,3,5,8,10,12,14]])
ax[1].set_xlim([0,6])
ax[1].set_xticks([0,1,2,3,4])
ax[1].set_xticklabels(context)
ax[1].set_title("SIM for all models with face feature")
ax[1].legend([model[i] for i in [1,3,5,8,10,12,14]], loc='best');
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5), sharey='row')

ax[0].plot(ig_table[:, [0,2,4,7,9,11,13,6]])
ax[0].set_xlim([0,6])
ax[0].set_xticks([0,1,2,3,4])
ax[0].set_xticklabels(context)
ax[0].set_title("IG for all models without face feature")
ax[0].legend([model[i] for i in [0,2,4,7,9,11,13,6]], loc='best');

ax[1].plot(ig_table[:, [1,3,5,8,10,12,14]])
ax[1].set_xlim([0,6])
ax[1].set_xticks([0,1,2,3,4])
ax[1].set_xticklabels(context)
ax[1].set_title("IG for all models with face feature")
ax[1].legend([model[i] for i in [1,3,5,8,10,12,14]], loc='best');
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5), sharey='row')

ax[0].plot(auc_table[:, [0,2,4,7,9,11,13,6]])
ax[0].set_xlim([0,6])
ax[0].set_xticks([0,1,2,3,4])
ax[0].set_xticklabels(context)
ax[0].set_title("AUC for all models without face feature")
ax[0].legend([model[i] for i in [0,2,4,7,9,11,13,6]], loc='best');

ax[1].plot(auc_table[:, [1,3,5,8,10,12,14]])
ax[1].set_xlim([0,6])
ax[1].set_xticks([0,1,2,3,4])
ax[1].set_xticklabels(context)
ax[1].set_title("AUC for all models with face feature")
ax[1].legend([model[i] for i in [1,3,5,8,10,12,14]], loc='best');
```
